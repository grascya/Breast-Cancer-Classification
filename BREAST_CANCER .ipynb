Most of the data are not normaly distributed with a skewness that tends to the right. we can also see a lot of outliers in the dataset

```python
#Target variable relationship
non_Cancerous = df[df["diagnosis"] =="B"]
Cancerous = df[df["diagnosis"] =="M"]

for col in df.select_dtypes("float"):
    plt.figure(figsize=(10,4))
    sns.distplot(non_Cancerous[col], label="non_Cancerous")
    sns.distplot(Cancerous[col], label="Cancerous")
    plt.legend()
```

According to distributions here are some Hypothesis:
- Cancerous tumors have highest radius lobes, outer perimeter of lobes, mean area of lobes
- Cancerous and non cancerous tumors have the same Surface Texture, Smoothness Levels, symmetry and fractal dimension
- Cancerous tumors are more compact and concave

```python
#correlation
plt.figure(figsize=(18,12))
mask = np.triu(np.ones_like(df.select_dtypes(include=['int','float']).corr(), dtype=bool))
sns.heatmap(df.select_dtypes(include=['int','float']).corr(), mask = mask, center=0, cmap ="RdBu", linewidths=1, annot=True, fmt=".2f", vmin=-1, vmax=1)
plt.show()
```

Radius_mean, area_mean, perimeter_mean, compactness_mean, concave points_mean, concavity_mean have strong correlations

```python
#hypothesis testing
#here we are using Wilcoxon-Mann-Whitney test because the data is not normaly distributed
from scipy.stats import mannwhitneyu

def u_test(col):
    alpha =0.05
    stat, p = mannwhitneyu(Cancerous[col],non_Cancerous[col])
    if p < alpha:
        return 'H0 Rejected'
    else:
        return 'Failed to reject H0'
    
```

```python
for col in df.select_dtypes("float"):
    print(f'{col:-<50}{u_test(col)}')
```

Because there is a statistically significant difference between Cancerous tumors and non_Cancerous tumors, we can say in relation to what we have seen on the graph that:
- Cancerous tumors have highest mean radius lobes, mean outer perimeter of lobes, mean area of lobes
- Cancerous tumors are more compact and concave 
- the null hypothesis could not be rejected so we can say that there is no difference between the mean of fractal dimension, the texture standard error and the smoothness standard error of the cancerous and the non cancerous tumors.

# PREPROCESSING

```python
#import librairies for data preprocessing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import make_pipeline
```

```python
#put the target variable into a numeric format
df["diagnosis"] = df["diagnosis"].replace(["M","B"],(1,0))
```

```python
#feature slection
#because some features are strongly statistically correlated with other features we can remove then. here we want to remove features that have strong correlation 0.9 for example with two or more features in order to reduce noice
#id variable seems to be irrelevant
new_df = df.drop(["id","area_mean", "area_se", "radius_se", "radius_worst","area_worst","perimeter_worst"], axis=1)
new_df.head()
```

```python
X=new_df.drop("diagnosis", axis=1)
y=new_df['diagnosis']
```

```python
# split the data into training and test set
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2, stratify=y, random_state=142)
ytrain
```

```python
#data standardisation
scaler = MinMaxScaler()
xtrain_scaled = scaler.fit_transform(xtrain)
xtest_scaled = scaler.transform(xtest)
```

# MODELLING

```python
#import librairies for modelling
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression

#librairies for hyperparameters tunning
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

#librairies for model evaluation
from sklearn.model_selection import learning_curve
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score
```

```python
# model instatiation

rdfc = RandomForestClassifier(random_state=142)
ada = AdaBoostClassifier(random_state=142)
tree = DecisionTreeClassifier(random_state=142)
svm = SVC(random_state=142, kernel="linear")
neighbors = KNeighborsClassifier()
logis = LogisticRegression(random_state=142)
```

```python
#dictionnaries 
dicts = {
    'RandomForestClassifier':rdfc,
    'AdaBoostClassifier':ada,
    'DecisionTreeClassifier':tree,
    'SVM':svm,
    'KNeighborsClassifier':neighbors,
    'LogisticRegression': logis
} 
```

```python
#function to evaluate our model 

def evaluation(model):
    model.fit(xtrain_scaled, ytrain)
    print(model.score(xtest_scaled, ytest))
    ypred = model.predict(xtest_scaled)
    print(confusion_matrix(ytest, ypred))
    print(classification_report(ytest, ypred))
    
    N, train_score, val_score = learning_curve(model, xtrain_scaled, ytrain, cv =5, scoring='precision', train_sizes = np.linspace(0.1,1,10))
    plt.figure(figsize=(12,8))
    plt.plot(N, train_score.mean(axis=1), label ="Training score")
    plt.plot(N, val_score.mean(axis=1), label = "Validation score")
    plt.legend()
```

```python
for name, model in dicts.items():
    print(name)
    evaluation(model)
```

SVM and LogisticRegression performed well, we have an accuracy of 97.36 and a f1_score of 96. Now let's try to tune the least good model: the DecisionTreeClassifer

```python
tree.get_params()
```

```python
#parameters
param_grid = {
    'max_depth': [2, 4, 6, 8],
    'min_samples_split': [2, 4, 6, 8],
    'min_samples_leaf': [1, 2, 4],
    'max_features': [None, 'sqrt', 'log2'],
    'criterion': ['gini', 'entropy']
}
```

```python
#hyperparameter tunning
grid_search = GridSearchCV(tree, param_grid, cv=5, )
grid_search.fit(xtrain_scaled, ytrain)
print(grid_search.best_params_)
y_pred = grid_search.predict(xtest_scaled)
print(classification_report(ytest, y_pred))
```

After tunning the DecisionTreeClassifer we have better accuracy parameters shown below 

```python
#let's evaluate our new model
evaluation(grid_search.best_estimator_)
```
